```python
import requests
from bs4 import BeautifulSoup

base_url = "https://books.toscrape.com/catalogue/page-{}.html"
books = []

for page in range(1, 10):  # scrape first 5 pages for now
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for article in soup.find_all('article', class_='product_pod'):
        title = article.h3.a['title']
        price = article.find('p', class_='price_color').text.strip()
        availability = article.find('p', class_='instock availability').text.strip()
        rating = article.p['class'][1]  # rating is hidden as a class name
        link = "https://books.toscrape.com/catalogue/" + article.h3.a['href']
        
        books.append({
            'title': title,
            'price': price,
            'availability': availability,
            'rating': rating,
            'link': link
        })

print(books[:3])  # preview first 3 results
```

```python
import pandas as pd

df = pd.DataFrame(books)
df.to_csv('books_data.csv', index=False)
print("Data saved to books_data.csv")
```

```python
import matplotlib.pyplot as plt

df['rating'].value_counts().plot(kind='bar')
plt.title('Number of Books by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
